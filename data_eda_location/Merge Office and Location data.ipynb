```python
import pandas as pd
```

```python
office_data = pd.read_excel('../data/rental_office_(july_to_nov_2018).xlsx')
office_data.head()
```

```python
location_data = pd.read_csv('../data/office_info_clean.csv')
location_data.head()
```

```python
location_data['location'] = list(zip(location_data.latitude, location_data.longitude))
location_data.head()
```

```python
locationDict = dict(zip(location_data.code, location_data.location))
locationDict
```

```python
office_data['location_map'] = office_data['code'].map(locationDict)
```

```python
office_data.head()
```

```python
office_data[['latitude','logitude']] = pd.DataFrame(office_data.location_map.tolist(), index= office_data.index)
office_data.head()
```

```python
office_data.to_excel("Rental_Office_July_to_Nov_2018.xlsx")
```
