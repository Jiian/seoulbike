```python
import pandas as pd
```

```python
#Read the dataset
df_seoul = pd.read_csv('../data/bike_clean.csv')
df_seoul.head()
```

```python
#Info and data type of each column of the datset
df_seoul.info()
```

```python
#Change type of date column to datetime
df_seoul['datetime'] =  pd.to_datetime(df_seoul['datetime'])
df_seoul['date'] = pd.to_datetime(df_seoul['date'])
df_seoul['month'] = df_seoul.datetime.dt.month
df_seoul['week'] = df_seoul.datetime.dt.isocalendar().week # week number
df_seoul['day_of_month'] = df_seoul.datetime.dt.day # 1 to 31
df_seoul['day_of_year'] = df_seoul.datetime.dt.dayofyear # 1 to 365
df_seoul['day_of_week'] = df_seoul.datetime.dt.dayofweek # mon, tue ... 
df_seoul['dayName_of_week'] = df_seoul.date.dt.day_name()
df_seoul.info()
```
